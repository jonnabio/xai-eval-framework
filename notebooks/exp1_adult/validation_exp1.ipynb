# Experiment 1 Validation: RF+SHAP vs XGBoost+LIME

This notebook validates the results of Experiment 1 (Adult Dataset). We analyze:
1. **Model Performance**: Accuracy, ROC-AUC.
2. **Explanation Quality**: Fidelity, Stability, Sparsity, Cost.
3. **Consistency**: Do the explanations make sense?


```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_theme(style="whitegrid")
```

```python
# Load Results
rf_shap_path = Path("../../experiments/exp1_adult/results/rf_shap/results.json")
xgb_lime_path = Path("../../experiments/exp1_adult/results/xgb_lime/results.json")

def load_results(path):
    if not path.exists():
        print(f"Warning: {path} does not exist.")
        return None
    with open(path, 'r') as f:
        return json.load(f)

rf_shap = load_results(rf_shap_path)
xgb_lime = load_results(xgb_lime_path)
```

```python
# 1. Model Performance Check
if rf_shap:
    print("RF Model Info:", rf_shap.get('model_info'))

if xgb_lime:
    print("XGB Model Info:", xgb_lime.get('model_info'))
```

```python
# 2. Explanation Metrics comparison
metrics = []

if rf_shap:
    agg = rf_shap.get('aggregate_metrics', {})
    agg['Method'] = 'RF+SHAP'
    metrics.append(agg)

if xgb_lime:
    agg = xgb_lime.get('aggregate_metrics', {})
    agg['Method'] = 'XGB+LIME'
    metrics.append(agg)

df_metrics = pd.DataFrame(metrics)
print(df_metrics[['Method', 'fidelity', 'stability', 'sparsity', 'cost']])
```

```python
if not df_metrics.empty:
    df_melt = df_metrics.melt(id_vars=['Method'], value_vars=['fidelity', 'stability', 'sparsity'])
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_melt, x='variable', y='value', hue='Method')
    plt.title("Explanation Metrics Comparison")
    plt.show()
```
